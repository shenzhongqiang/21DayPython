# 自定义函数

```python
# 使用关键字def
# 括号里的num是参数
# return用于从函数返回值
def fibs(num):
    fibs = []
    for i in range(num):
        if i == 0:
            fibs.append(0)
        elif i == 1:
            fibs.append(1)
        else:
            fibs.append(fibs[-2] + fibs[-1])
    return fibs
fibs(12)
```

# 函数参数

```python
# 函数也可以不接受任何参数，不返回任何值
def print_hello():
    print("hello")
```

```python
# 函数也可以接受列表作为参数
def calc_sum(numbers):
    result = 0
    for i in numbers:
        result += i
    return result
calc_sum([1,2,3])
```

```python
# 函数也可以接受字典作为参数
words = {"the": 10, "you": 5, "name":1}
def count_words(words):
    result = 0
    for k, v in words.items():
        result += v
    return result
count_words(words)
```

```python
# 可变参数与不可变参数
# 不可变参数 - 字符串、数和元组
def change_value(x):
    print("in change_value: %d" % x)
    x = 2
    
a = 1
change_value(a)
print("in main: %d" % a)
```

```python
a=1
b=a
b+=1
print(a)
```

```python
# 可变参数 - 列表、字典的值
def change_value(array):
    array[0] += 1

array = [1, 2, 3]
change_value(array)
print(array)
```

```python
def change_value(person):
    person["age"] = 30

person = {"name": "john", "age": 25}
change_value(person)
print(person)
```

```python
def print_person(name, age):
    message = "{} is {} year-old".format(name, age)
    print(message)

# 使用位置参数调用
print_person("john", 27)

# 使用关键字参数调用
print_person(name="john", age=27)
print_person(age=27, name="john")
```

```python
# 参数默认值
def print_person(name="jack", age=28):
    message = "{} is {} year-old".format(name, age)
    print(message)

print_person()
print_person(name="henry")
```

# 作用域

```python

# 全局作用域
x = 1

# 内部作用域
def change_value():
    x = 10
    print("in change_value:", x)

change_value()
print("in main:", x)
```

```python
# 在函数内部使用全局变量
x = 1
def change_value():
    global x
    x = 42
    print("in change_value:", x)
change_value()
print("in main:", x)
```

# 递归

```python
# 什么是递归 - 函数调用自身
# 两个关键点 - 什么时候递归调用，什么时候结束调用
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)
fib(10)
```

# 实战项目答案

```python
size = 8
array = [[0] * size for i in range(size)]

def check(row, col):
    for i in range(size):
        if array[i][col] == 1:
            return False

    i = row - 1
    j = col - 1
    while i >= 0 and j >= 0:
        if array[i][j] == 1:
            return False
        i -= 1
        j -= 1

    i = row - 1
    j = col + 1
    while i >= 0 and j <= size-1:
        if array[i][j] == 1:
            return False
        i -= 1
        j += 1
    return True

def print_result():
    for row in range(size):
        for col in range(size):
            if array[row][col] == 1:
                print("*", end="")
            else:
                print("o", end="")
        print()
    print()

def find_queen(row):
    if row > size-1:
        print_result()
        return
    for col in range(size):
        if check(row, col):
            array[row][col] = 1
            find_queen(row+1)
            array[row][col] = 0

if __name__ == "__main__":
    find_queen(0)
```
